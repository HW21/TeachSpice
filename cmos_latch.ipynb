```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
from spice import Circuit, DcOp, Resistor, Mos, Diode
from spice.analysis import Contour
from spice.tests.test_cmos_latch import CmosLatch
```

```python
c = Contour(CmosLatch())
c.explore(xmin=-0.5, xmax=1.5, xstep=0.1)
```

```python
xs = np.transpose(c.xs)
ys = np.transpose(c.ys)
dxs = np.transpose(c.dxs)
xfs = np.transpose(c.xfs)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs[0], xs[1], xfs[1])
ax.view_init(azim=30)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(xs[0], xs[1], np.sqrt(ys[0]**2+ys[1]**2), color='red')
ax.view_init(azim=90)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(xs[0], xs[1], ys[0], alpha=0.5)
ax.plot_trisurf(xs[0], xs[1], ys[1], alpha=0.5)
ax.view_init(azim=75)
```
