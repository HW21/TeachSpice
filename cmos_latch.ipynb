```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
from spice import Circuit, DcOp, Resistor, Mos, Diode
from spice.analysis import Contour
from spice.tests.test_cmos_latch import CmosLatch
```

```python
c = Contour(CmosLatch())
c.explore(xmin=-0.5, xmax=1.5, xstep=0.1)
```

```python
xs = np.transpose(c.xs)
ys = np.transpose(c.ys)
dxs = np.transpose(c.dxs)
xfs = np.transpose(c.xfs)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs[0], xs[1], xfs[1])
ax.view_init(30,240)
ax.set_xlabel("Initial Guess V(a)")
ax.set_ylabel("Initial Guess V(b)")
ax.set_zlabel("Solution V(b)")
```

```python
def surf():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    line = ax.plot_trisurf(xs[0], xs[1], np.sqrt(ys[0]**2+ys[1]**2), cmap=plt.cm.viridis)
    ax.view_init(30,70)
    ax.set_zticklabels([])
    ax.set_xlabel("Guess V(a)")
    ax.set_ylabel("Guess V(b)")
    return fig, ax, line

fig, ax, line = surf()
```

```python
from matplotlib.animation import FuncAnimation

fig, ax, line = surf()

def update(i):
    ax.view_init(30, i)
    return line, ax

anim = FuncAnimation(fig, update, frames=np.arange(0, 180, 10), interval=150)
anim.save('cmos-latch-anim.gif')
```
