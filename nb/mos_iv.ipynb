```python
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 8]
```

```python
from spice import Circuit, DcOp, Resistor, Mos, Diode
from spice.analysis import Contour
```

```python
from spice.components import Mos
```

```python
p = Mos(polarity=-1, conns={})

for g in range(11):
    vdss = []
    ops = []
    for k in range(101):
        vds = k * 0.01
        ops.append(p.op_point({'s':0,'b':0,'d':-1*vds,'g':-0.1*g}))
        vdss.append(vds)

    plt.plot(vdss, [op['ids'] for op in ops], label=g)
plt.legend()
# plt.yscale('log')
```

```python
n = Mos(polarity=+1, conns={})

for g in range(11):
    vdss = []
    ops = []
    for k in range(101):
        vds = k * 0.01
        ops.append(n.op_point({'s':0,'b':0,'d':vds,'g':0.1*g}))
        vdss.append(vds)

    plt.plot(vdss, [op['gm'] for op in ops], label=g)
plt.legend()
plt.yscale('log')
```
