```python
import scipy.optimize
```

```python
def f(x):
    #     print(f'x={x}')
    i1 = (1.0-x) / 10e3
    i2 = x / 10e6
    rv = (i1-i2)**2
    #     print(f'rv={rv}')
    return rv
```

```python
options = dict(disp=True)
scipy.optimize.minimize(f, [0], options=options, method='nelder-mead')
```

```python
import matplotlib.pyplot as plt
```

```python
x = [0.01*k - 1.0 for k in range(300)]
y = list(map(f, x))
```

```python
plt.plot(x,y)
```
