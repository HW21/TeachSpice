```python
from spice import Circuit, DcOp, Resistor, Mos, Diode
from spice.analysis import Contour
from spice.tests.test_cmos_latch import CmosLatch
```

```python
c = Contour(CmosLatch())
c.explore(xmin=0, xmax=1.0, xstep=0.1)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
len(np.reshape(c.xs,(2,len(c.xs)))[0])
```

```python
xs = np.transpose(c.xs)
ys = np.transpose(c.ys)
dxs = np.transpose(c.dxs)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs[0], xs[1], np.sqrt(ys[0]**2+ys[1]**2))
ax.view_init(azim=-150)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs[0], xs[1], np.sqrt(dxs[0]**2+dxs[1]**2))
ax.view_init(azim=120)
```
